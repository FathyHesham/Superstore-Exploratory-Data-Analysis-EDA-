#### üìÖ Converting Integer Date Format to MM/DD/YYYY

```python
'''
output ==> 6 0 28 0 2016 ==>  conver to ==> 6 / 28 / 2016

    * 6 => month
    * 28 => day
    * 2016 => year

    3 case : 
        - len 10 ==> 11 0 11 0 2016
        - len 9  ==> 6  0 14 0 2014
        - len 8  ==> 9  0  1 0 2014
'''

# Function to Convert Integer Date Format to MM/DD/YYYY
def convert_integer_date(date_int):
    date_str = str(date_int)  # Convert the integer date to a string
    year = date_str[-4:]      # Extract the year from the last four characters of the string
    if len(date_str) == 10:   # Check if the string length is 10
        month = date_str[:-8] # Extract the month from the string
        day = date_str[-7:-5] # Extract the day from the string
    elif len(date_str) == 9:  # Check if the string length is 9
        month = date_str[0]   # Extract the month from the first character of the string
        day = date_str[-7:-5] # Extract the day from the string
    elif len(date_str) == 8:  # Check if the string length is 8
        month = date_str[0]   # Extract the month from the first character of the string
        day = date_str[1:3]   # Extract the day from the second and third characters of the string
    else:
        raise ValueError("Invalid date format")  # Raise an error for invalid date formats
    formatted_date = f"{month}/{day}/{year}"  # Format the date as MM/DD/YYYY
    return formatted_date  # Return the formatted date

# Applying the Conversion Function to the 'Ship Date' Column
orders_dataset['Ship Date'] = orders_dataset['Ship Date'].apply(convert_integer_date)

# Displaying the First 5 Rows of the 'Ship Date' Column
orders_dataset["Ship Date"].head(5)
```

#### üåçüèôÔ∏è Splitting the 'Location' Column into 'Country' and 'City'

```python
# Splitting the 'Location' column into 'Country' and 'City' columns:
# Using str.split() with expand=True splits the 'Location' column by commas,
# creating two new columns 'Country' and 'City' in the orders_dataset DataFrame.
orders_dataset[['Country', 'City']] = orders_dataset['Location'].str.split(',', expand=True)

# Reordering the columns to place 'Country' and 'City' after 'Location':
# Convert the column names to a list and find the index of 'Location'.
columns = list(orders_dataset.columns)
location_idx = columns.index('Location')

# Define the new order of columns by placing 'Country' and 'City' after 'Location'.
new_columns_order = columns[:location_idx + 1] + ['Country', 'City'] + columns[location_idx + 1:-2]

# Apply the new column order to the orders_dataset DataFrame:
# Update orders_dataset with the new column order to reflect the split and reordering.
orders_dataset = orders_dataset[new_columns_order]

# Drop the original 'Location' column if it's no longer needed:
# Remove the 'Location' column from orders_dataset since it's been split into 'Country' and 'City'.
orders_dataset = orders_dataset.drop(columns=['Location'])

# Print the updated DataFrame to verify the changes:
# Display the first few rows of orders_dataset to confirm the split, reordering,
# and removal of the 'Location' column.
orders_dataset.head(1)
```

# <p style="font-family:newtimeroman;font-size:150%;text-align:center;color:red;">Feature Engineering üöÄüî•</p>

#### Calculate Discount Value and COGS üìâüí∞

```python
# Calculate Discount Value:
# Multiply the 'Sales' column by -(Discount) to compute the Discount Value,
# ensuring it is negative regardless of the original sign of Discount.
orders_dataset["Discount_Value"] = orders_dataset["Sales"] * (-(orders_dataset["Discount"]))

# Calculate COGS (Cost Of Goods Sold):
# Compute COGS by subtracting the sum of 'Sales' and 'Disount_Value' from 'Profit',
# and then negating the result. This calculation represents the cost associated
# with goods sold, considering discounts and profit.
orders_dataset["COGS"] = (-((orders_dataset["Sales"] + orders_dataset["Discount_Value"]) - orders_dataset["Profit"]))

# Reorder columns to place Discount Value and COGS after Discount:
# Retrieve the current column names and find the index of 'Discount'.
# Define a new column order that inserts 'Disount_Value' and 'COGS'
# after 'Discount', moving 'Profit' to the end of the DataFrame.
columns = list(orders_dataset.columns)
discount_idx = columns.index('Discount')
new_columns_order = columns[:discount_idx + 1] + ['Discount_Value', 'COGS'] + columns[discount_idx + 1:-3] + ['Profit']

# Apply the new columns order to the DataFrame:
# Update 'orders_dataset' with the new column order to reflect the changes made.
orders_dataset = orders_dataset[new_columns_order]

# Display the first few rows of the updated dataset:
# Show the DataFrame to verify the calculations and column reordering.
orders_dataset.head()
```

```python
# Assuming 'Ship Mode', 'Postal Code', 'COGS', and 'People' are columns in your DataFrame.
# Adjust the column names accordingly if they differ.

# Find the index of 'Ship Mode' and 'Postal Code' in the current columns list.
columns = list(orders_dataset.columns)
ship_mode_idx = columns.index('Ship Mode')
postal_code_idx = columns.index('Postal Code')
cogs_idx = columns.index('COGS')

# Define the new order of columns.
new_columns_order = (
    columns[:ship_mode_idx + 1] +  # Columns before 'Ship Mode'
    ['Returned'] +                 # Returned column after 'Ship Mode'
    columns[ship_mode_idx + 1:postal_code_idx + 1] +  # Columns between 'Ship Mode' and 'Postal Code'
    ['Person'] +                   # 'People' column after 'Postal Code'
    columns[postal_code_idx + 1:cogs_idx + 1] +  # Columns between 'Postal Code' and 'COGS'
    ['Shipping Cost Per Unit'] +            # 'Shipping Cost' column after 'COGS'
    columns[cogs_idx + 1:-3]       # Columns after 'Shipping Cost'
)

# Apply the new column order to the orders_dataset DataFrame.
orders_dataset = orders_dataset[new_columns_order]

# Rename 'Returned' column to 'Returned Order'.
orders_dataset = orders_dataset.rename(columns={'Returned': 'Returned Order'})

# Print the updated DataFrame to verify the changes.
orders_dataset.head(1)
```

```python
# Calculate Total Shipping based on Quantity and Shipping Cost Per Unit
orders_dataset["Total Shipping"] = orders_dataset["Quantity"] * orders_dataset["Shipping Cost Per Unit"]

# Assuming 'Shipping Cost Per Unit' is already in your DataFrame,
# find its index and insert 'Total Shipping' after it.
columns = list(orders_dataset.columns)
shipping_cost_per_unit_idx = columns.index('Shipping Cost Per Unit')

# Define the new order of columns with 'Total Shipping' inserted after 'Shipping Cost Per Unit'
new_columns_order = (
    columns[:shipping_cost_per_unit_idx + 1] +  # Columns before 'Shipping Cost Per Unit'
    ['Total Shipping'] +                       # 'Total Shipping' column after 'Shipping Cost Per Unit'
    columns[shipping_cost_per_unit_idx + 1:-1]    # Columns after 'Shipping Cost Per Unit'
)

# Apply the new column order to the orders_dataset DataFrame
orders_dataset = orders_dataset[new_columns_order]

# Print the updated DataFrame to verify the changes.
orders_dataset.head(1)
```

### Time Series Analysis

**25. How do sales, Discount_Value, Profit, COGS and Total Shipping trends change over time (monthly)?**

```python
# Assuming 'final_dataset' is your DataFrame and 'Order Date' is in datetime format
final_dataset['Order Date'] = pd.to_datetime(final_dataset['Order Date'])

# Aggregate data by month
monthly_data = final_dataset.groupby(pd.Grouper(key='Order Date', freq='ME')).agg({
    'Sales': 'sum',  # Calculate total sales for each monthE
    'Discount_Value': 'sum',  # Calculate total discount value for each month
    'Profit': 'sum',  # Calculate total profit for each month
    'COGS': 'sum',  # Calculate total cost of goods sold for each month
    'Total Shipping': 'sum'  # Calculate total shipping cost for each month
}).reset_index()  # Add 'Order Date' back as a column in the result

# Create a figure and axes with a grid layout (2 rows, 3 columns)
fig, axs = plt.subplots(2, 3, figsize=(18, 12))

# Plot Sales
axs[0, 0].plot(monthly_data['Order Date'], monthly_data['Sales'], marker='o', linestyle='-', color='blue')
axs[0, 0].set_ylabel('Sales ($)', fontsize=12)
axs[0, 0].set_title('Sales', fontsize=14, fontweight='bold')

# Plot Discount_Value
axs[0, 1].plot(monthly_data['Order Date'], monthly_data['Discount_Value'], marker='o', linestyle='-', color='orange')
axs[0, 1].set_ylabel('Discount Value ($)', fontsize=12)
axs[0, 1].set_title('Discount Value', fontsize=14, fontweight='bold')

# Plot Profit
axs[0, 2].plot(monthly_data['Order Date'], monthly_data['Profit'], marker='o', linestyle='-', color='green')
axs[0, 2].set_ylabel('Profit ($)', fontsize=12)
axs[0, 2].set_title('Profit', fontsize=14, fontweight='bold')

# Plot COGS
axs[1, 0].plot(monthly_data['Order Date'], monthly_data['COGS'], marker='o', linestyle='-', color='red')
axs[1, 0].set_ylabel('COGS ($)', fontsize=12)
axs[1, 0].set_title('COGS', fontsize=14, fontweight='bold')

# Plot Total Shipping
axs[1, 1].plot(monthly_data['Order Date'], monthly_data['Total Shipping'], marker='o', linestyle='-', color='purple')
axs[1, 1].set_ylabel('Total Shipping ($)', fontsize=12)
axs[1, 1].set_title('Total Shipping', fontsize=14, fontweight='bold')

# Remove the empty subplot (axs[1, 2])
fig.delaxes(axs[1, 2])

# Customize x-axis labels rotation for better readability
for ax in axs.flat:
    ax.tick_params(axis='x', rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
